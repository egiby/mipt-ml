**Вопрос 1**: Объясните, чем отличается k-nearrest Neighbours от k-weighted nearest neighbours.

Пусть алгоритм классификации имеет вид $a(u) = argmax_{y \in Y}({\Gamma_y(u)})$, где $\Gamma_y(u)$ - некая функция.

k-nearrest Neighbours - это если $\Gamma_y(u) = \sum_{i \in K}{[y(i) = y]}$, где $K$ - индексы ближайших $k$ соседей.

k-weighted nearest neighbours - если $\Gamma_y(u) = \sum_{i \in K}{[y(i) = y] \cdot w(i)}$, где w(i) - вес точки.

**Вопрос 2**: Как изменяется абсолютное расстояние между объектами выборки при изменении метрики минковского с $p=1$ до $p=\infty$?

В метрике Минковского $p(x, y) = \left(\sum_i{|x_i - y_i|^p}\right)^\frac{1}{p}$. Можно посмотреть, например, на расстояние между точками (0, 0) и (1, 1). Для p = 1 это 2, а для p = $\infty$ - 1. Можно показать, что в промежуточных значениях расстояние будет монотонно убывать, а в силу равенства $p(0, \alpha x) = \alpha p(0, x)$ рассмотрения этих двух точек достаточно. 

**Вопрос 3**: Поясните, в чем суть проклятия размерности?
  
При фиксированном числе точек чем больше размерность, тем больше разрежены данные. По сравнению с расстоянием до ближайших соседей различия в них получаются незначительны.

** Вопрос 4**: Что такое метрический отступ?

Пусть алгоритм классификации имеет вид $a(u) = arg max_{y \in Y}({\Gamma_y(u)})$. Тогда *метрический отступ (margin)* -- это функция $M(x_i) = \Gamma_{y_i}(x_i) - max_{y \in Y \setminus y_i}{(\Gamma_y(x_i))}$ 

** Вопрос 5**: На какие типы можно разделить объекты обучающей выборки с точки зрения значения значения метрического отступа?
Какие объекты стоит исключить из выборки?

Бывают *эталонные(с большим отступом)*, *неинформативные(с меньшим, но положительным отступом)*, *пограничные(около нуля)*, *ошибочные(отрицательный отступ)* и *шумы(наибольший по модулю отрицательный отступ)*. Стоит выбрасывать именно шумы.

**Вопрос 6**:  Что такое функционал эмпирического риска? Приведите примеры.

Это ожидаемая величина потери: $R(a) = \sum_{y \in Y}\sum_{s \in Y}{[y \neq s] P_y P(a = s|y)}$

**Вопрос 7**:  В чём суть явления переобучения?

Алгоритм слишком "приспосабливается" к ответам обучающей выборки. Примером может быть попытка приближения функции $f(x) = C$ по известным значениям $f(x_i) + \varepsilon$ (где $\varepsilon$ - некий шум) многочленом по известным точкам. На обучающей выборке ответ будет абсолютно точным, а вот на точках с большим по модулю $x$ значения будут приближаться крайне плохо. 

**Вопрос 8**: Напишите формулу для complete cross-validation.

Пусть $(x_i, y_i)_{i = 1}^{n}$ - обучающая выборка, хотим описать complete cross-validation по k элементам.
Пусть $\Gamma(A)$ - алгоритим $\Gamma$, обученный на выборке $A$. Тогда:

$cv = \frac{1}{C_{|X|}^k \cdot (|X| - k)} \sum_{(A \subset X, |A| = k)}{\sum_{x_i \in X \setminus A}{[\Gamma(A)(x_i) = y_i]}}$

```python
import warnings
import numpy as np
import pandas as pd
from collections import Counter

warnings.simplefilter("ignore")


class kNNClassifier():
    def __init__(self, n_neighbours, metric):
        """
        Parameters
        ----------
        n_neighbours: int
            Число соседей

        metric: *alias
            метрика измерения расстояний

        """
        
        # Тут было что-то странное. Понятно, что видимо хотелось передавать 
        # estimator из sklearn, но тогда зачем в predict предлагалось считать расстояние руками?
        self.n_neighbours = n_neighbours
        self.metric = metric
        
    def fit(self, X, y):
        """
        
        Parameters
        ----------
        X: 2d np.array
        y: 1d np.array
        """

        # Тут храните описание объектов обучающей выборки
        self.X_learn = np.array(X)

        # Тут храните ответы по каждому объекту обучающей выборки
        self.y_learn = np.array(y)
        
        return self
    
    def _get_color(self, colors):
        # most_common возвращает список из одной пары (key, value)
        return Counter(colors).most_common(1)[0][0]
    
    def predict(self, X):
        """

        Parameters
        ----------
        X: 2d np.array, матрица объекты - признаки на которых нужно сказать ответ

        Returns
        -------
        y_pred: 1d np.array, Вектор классов для каждого объекта
        """
        
        dist = [[-1] * len(self.X_learn) for _ in xrange(len(X))]
        for i in xrange(len(X)):
            # =======================================
            # рассчитайте расстояние до каждого объекта обучающей выборки
            # ======================================
            for j in xrange(len(self.X_learn)):
                dist[i][j]= self.metric(X[i], self.X_learn[j])
    
        # =======================================
        # предскажите класс каждого из объектов
        # =======================================
        y_pred = [self._get_color(np.array(list(sorted(zip(d, self.y_learn)))[:self.n_neighbours])[:,1])
                  for d in dist]
        
        return y_pred
```

Размер обучающей выборки $l$ взять таким, чтобы можно было значения CV при $k=10\cdot l$.

В качестве алгоритма классификации можно взять

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import train_test_split

L = 500

def get_sample(n):
    random.shuffle(indices)
    sample = train.ix[indices[:n]]
    
    return np.asarray(sample[range(1, train.shape[1])]), np.asarray(sample[[0]]).ravel()

def test_sample(k, n=5):
    X, y = get_sample(k + L)
    
    cv = .0
    
    for i in range(n):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=k)

        clf = KNeighborsClassifier().fit(X_train, y_train)
        
        cv += accuracy_score(clf.predict(X_test), y_test)
    
    return cv / n
    
```

```python
from sklearn.neighbors import NearestNeighbors
import pylab

# Да, это функция рассчета margine из семинара
def calculate_margin(estimator, class_id, example, Y_train, debug=False):
    """
    We assume that we have uniform weights. We also assume
    that we fit our estimator on the whole set. It means
    that we already have our example inside estimator. That is
    why, the first nearest neighbor would be example object.
    
    @param estimator: KNeighborsMixin sublcassed object
        should be already fitted.
    @param class_id: example object has class class_id
    @param example: numpy.array feature represenation of
        example (1x784 matrix based on 28x28 reshaping)
    @param Y_train: we should know object class by its index id
    @params debug: do we need to print debug information?
    """
    counts = Counter(Y_train[estimator.kneighbors([example], return_distance=False)[0][1:]])
    if debug:
        # in production environment we should use logging module!
        print(counts)
    margin = counts[class_id]
    del counts[class_id]
    if counts:
        margin -= max(counts.values())
    margin *= 1.0 / (estimator.n_neighbors - 1)
    return margin

# И это тоже функция из семинара
def collect_margins(n_neighbors, margin_calculator, X_train, Y_train, 
                    debug_example_id = None, **kwargs):
    margins = []
    estimator = NearestNeighbors(n_neighbors=n_neighbors + 1, **kwargs)
    estimator.fit(X_train)
    for example_id, (class_id, example) in enumerate(zip(Y_train, X_train)):
        margin = margin_calculator(estimator, class_id, example, Y_train, 
                                   debug=(example_id == debug_example_id))
        margins.append(margin)

    if debug_example_id is not None:
        print(margins[debug_example_id])
    return margins
```

```python
def test_metric(n_neighbors, X_train, y_train, cosine_margins, metric, algo, **kwargs):
    margins = collect_margins(n_neighbors, calculate_margin, X_train, y_train, 
                              metric=metric, algorithm=algo, **kwargs)
    plt.plot(range(len(margins)), sorted(margins), label=metric)
    plt.plot(range(len(cosine_margins)), sorted(cosine_margins), label='cosine')
```

```python
N = 1000
k = 5
X_train, y_train = get_sample(N)
print "sample finished"
cosine_margins = collect_margins(k, calculate_margin, X_train, y_train, 
                              metric='cosine', algorithm='brute')
```

```python
metrics = ['cityblock', 'euclidean', 'l1', 'l2', 'chebyshev', 'correlation']
algos = ['kd_tree', 'kd_tree', 'ball_tree', 'ball_tree', 'kd_tree', 'brute']

for algo, metric in zip(algos, metrics):
    test_metric(k, X_train, y_train, cosine_margins, metric=metric, algo=algo)
    plt.legend(loc='upper left')
    plt.show()
```

Видно, что стоит отдельно посмотреть метрику *correlation*. 

```python
N = 5000
X_train, y_train = get_sample(N)
print "sample finished"
cosine_margins = collect_margins(k, calculate_margin, X_train, y_train, 
                              metric='cosine', algorithm='brute')

plt.figure(figsize=(16, 10))
test_metric(k, X_train, y_train, cosine_margins, metric='correlation', algo='brute')
plt.legend()
plt.show()
```
